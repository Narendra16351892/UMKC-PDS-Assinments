```python
import pandas as pd
```

```python
df = pd.read_csv('/content/data_clean/StudentsPerformance.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1.Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='gender', y='math_score')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.title('Violin Plot of Math Scores by Gender')
plt.grid(True)
plt.show()

# 2.Swarm Plot
plt.figure(figsize=(12, 6))
sns.swarmplot(data=df, x='parental_level_of_education', y='reading_score')
plt.xlabel('Parental Level of Education')
plt.ylabel('Reading Score')
plt.title('Swarm Plot of Reading Scores by Parental Level of Education')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()


# 3. PairGrid
g = sns.PairGrid(df[['math_score', 'reading_score', 'writing_score']])
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
plt.suptitle('PairGrid of Math, Reading, and Writing Scores')
plt.show()






```

```python
# Joint Plot
sns.jointplot(data=df, x='reading_score', y='writing_score', kind='hex')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.title('Joint Plot of Reading vs. Writing Scores')
plt.show()
```

```python
# Heatmap
corr = df[['math_score', 'reading_score', 'writing_score']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
# Box Plot with Swarm Overlay
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='parental_level_of_education', y='math_score')
sns.swarmplot(data=df, x='parental_level_of_education', y='math_score', color='black', alpha=0.5)
plt.xlabel('Parental Level of Education')
plt.ylabel('Math Score')
plt.title('Box Plot with Swarm Overlay of Math Scores by Parental Level of Education')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
